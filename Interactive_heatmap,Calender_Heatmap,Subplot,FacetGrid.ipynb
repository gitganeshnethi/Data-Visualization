```python
import calmap
from ipywidgets import widgets,interact
from matplotlib import pyplot as plt
import seaborn as sb
import pandas as pd
import numpy as np
from pydataset import data
plt.rcParams['figure.figsize'] = [18,8]
```

```python
data = pd.read_excel('D:\\Data Visualization\\Datasets\\UN_Data.xlsx')
data.head()
data.shape
data1 = data[data['Month']!='Total']
data1.head()
data1 = data1[(data1.Reliability == 'Final figure, complete')]
main_data = data1[['Year','Month','Value','Country']]
main_data.head()
main_data.shape


country = widgets.Dropdown(options = ['All']+list(main_data.Country.unique()),value = 'All',description = 'Country')
country

```

```python
def countryjoin(x):

    if country.value != 'All':

        req_data = main_data[main_data['Country'] == country.value]

    else:

        req_data = main_data[['Year','Month','Value']].groupby(['Year','Month']).sum().reset_index()
    
    required = req_data.pivot('Month','Year','Value')
    
    index_order = ['January','February','March','April','May','June','July','August','September','October','November','December']
    reuired_new = required.reindex(index_order)
    sb.heatmap(reuired_new,cmap='hot_r',annot=True,fmt='.1f',linewidth=.4)
    
    plt.show()
    
interact(countryjoin,x = country)
```
