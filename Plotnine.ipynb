```python
import pandas as pd
import numpy as np
from pandas.api.types import CategoricalDtype
from plotnine import *
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
census_data = pd.read_csv("adult.csv")
```

```python
census_data.columns
```

```python
(ggplot(census_data)        
 + aes(x='sex',fill='marital.status') 
 + geom_bar(position='dodge')
 + labs(x = 'Gender', title = 'Gender w.r.to Marital Status')) 
```

```python
(ggplot(census_data)        
 + aes(x='age',fill='sex') 
 + geom_histogram(position='dodge')
 + labs(x = 'Age', title = 'Distribution of Age w.r.to Gender')) 
```

```python
(ggplot(census_data)        
 + aes(x='age', y ='hours.per.week',color='sex') 
 + geom_point(alpha=0.4))

```

```python
(ggplot(census_data)        
 + aes(x='age', y ='hours.per.week',color='income') 
 + geom_point(alpha=0.4)) + facet_wrap("workclass")

```

```python
import plotnine as pp
pp.options.figure_size=(20,10)
```

```python
(ggplot(census_data)        
 + aes(x='occupation', y ='age',fill='sex') 
 + geom_boxplot()
 + labs(x = 'Different Occupations', title = 'Gender Wise Distribution of Age w.r.to their Occupation')) 
```

```python
(ggplot(census_data)        
 + aes(x='age', y ='hours.per.week',color='income') 
 + geom_point(alpha=0.4)) + facet_wrap("native.country")
```

```python
(ggplot(census_data)        
 + aes(x='sex', y ='age',fill='income') 
 + geom_jitter()
 + geom_boxplot(alpha=0.2)
 + labs(x = 'Different Occupations', title = 'Gender Wise Distribution of Age w.r.to their Occupation')) 
```
